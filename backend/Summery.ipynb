```python
import spacy
import pytextrank
import sys
# Load the spaCy model
nlp = spacy.load("en_core_web_md")

# Add the TextRank component to the pipeline
nlp.add_pipe("textrank", last=True)

# Define a function for basic classification based on keywords
def classify_text(text):
    categories = {
        'sports': ['game', 'team', 'score', 'match', 'player'],
        'politics': ['government', 'election', 'party', 'minister', 'policy'],
        'crime': ['crime', 'murder', 'theft', 'robbery', 'arrest'],
        'economy': ['economy', 'market', 'investment', 'business', 'company']
    }
    
    scores = {category: 0 for category in categories}
    for category, keywords in categories.items():
        for keyword in keywords:
            if keyword in text.lower():
                scores[category] += 1
    
    return max(scores, key=scores.get)

# Function to generate summary
def generate_summary(text):
    doc = nlp(text)
    summary = "\n".join([sent.text for sent in doc._.textrank.summary(limit_phrases=15, limit_sentences=5)])
    return summary

# Prompt user to enter text
text = input("Enter the text for classification and summarization:\n")

# Classify the text
news_type = classify_text(text)
print(f"\nClassified News Type: {news_type}")

# List all available types
available_types = set([classify_text(text) for text in [text]])
print("\nAvailable News Types:")
for t in available_types:
    print(t)

# Prompt user to select a news type
selected_type = input("\nSelect a news type from the available options:\n")

if selected_type in available_types:
    # Generate and print summary for the selected type
    print("\nSummary for selected news type:")
    print(generate_summary(text))
else:
    print("Selected news type is not available.")
```
