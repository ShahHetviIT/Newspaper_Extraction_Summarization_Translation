```python
stop_words_ = set(stopwords.words('english'))
my_sw = [
    'make', 'amp', 'news', 'new', 'time', 'u', 's', 'photos', 'get', 'say', 'said', 'also', 
    'go', 'goes', 'went', 'just', 'now', 'one', 'two', 'three', 'many', 'take', 'come', 'back', 
    'still', 'know', 'look', 'people', 'year', 'years', 'day', 'days', 'week', 'weeks', 'month', 
    'months', 'like', 'well', 'even', 'way', 'lot', 'see', 'things', 'thing', 'work', 'need', 
    'want', 'use', 'used', 'using', 'every', 'really', 'much', 'first', 'second', 'next', 'last'
]

stop_words_.update(my_sw)  # Merge with custom stop words
wn = WordNetLemmatizer() #converting the word in its base form ex running->run

def black_txt(token):
    return token not in stop_words_ and token not in string.punctuation and len(token) > 2

def clean_txt(text):
    text = re.sub(r"'", "", text)  # Remove apostrophes
    text = re.sub(r"(\\d|\\W)+", " ", text)  # Remove non-word characters and digits
    clean_text = [wn.lemmatize(word, pos="v") for word in word_tokenize(text.lower()) if black_txt(word)]
    return " ".join(clean_text)
```

```python
def subj_txt(text):
    return  TextBlob(text).sentiment[1]

def polarity_txt(text):
    return TextBlob(text).sentiment[0]

def len_text(text):
    if len(text.split())>0:
         return len(set(clean_txt(text).split()))/ len(text.split())
    else:
         return 0
```

```python
# Make a copy of the DataFrame to avoid modifying a slice
df_copy = df.copy()
```

```python
# Concatenate 'headline' and 'short_description' into 'text'
df_copy['text'] = df_copy['headline'] + " " + df_copy['short_description']

# Proceed with further processing
df_copy['text'] = df_copy['text'].swifter.apply(clean_txt)
df_copy['polarity'] = df_copy['text'].swifter.apply(polarity_txt)
df_copy['subjectivity'] = df_copy['text'].swifter.apply(subj_txt)
df_copy['len'] = df_copy['text'].swifter.apply(lambda x: len(x))
```

```python
min_class_count = 2
counts = df_copy['category'].value_counts()
rare_classes = counts[counts < min_class_count].index
df = df_copy[~df['category'].isin(rare_classes)]
X = df_copy[['text', 'polarity', 'subjectivity', 'len']]
y = df_copy['category']
v = dict(zip(list(y), df_copy['category'].to_list()))
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

text_clf = Pipeline([
    ('vect', CountVectorizer(
        analyzer="word",
        stop_words="english",
        ngram_range=(1, 3),     # Keeping trigrams for context capture (3 words)
        max_df=0.85,            # Exclude overly common terms appearing in 85%+ of docs
        min_df=3,               # Exclude terms that appear in fewer than 3 documents
        max_features=20000      # Increase feature count for richer representation
    )),
    ('tfidf', TfidfTransformer(
        use_idf=True,           # Use IDF to give more importance to rare terms
        smooth_idf=True         # Smooth IDF to avoid division by zero
    )),
    ('clf', MultinomialNB(
        alpha=0.1               # Reduce smoothing for more sensitivity to data
    )),
])
```

```python
text_clf.fit(x_train['text'].to_list(), list(y_train))
```

```python
X_TEST = x_test['text'].to_list()
Y_TEST = list(y_test)
```

```python
predicted = text_clf.predict(X_TEST)
```

```python
c = 0

for doc, category in zip(X_TEST, predicted):
    
    if c ==3:break
    
    print("-"*55)
    print(doc)
    print(v[category])
    print("-"*55)

    c = c + 1 
```

```python
from sklearn.metrics import accuracy_score

# Calculate accuracy
accuracy = accuracy_score(Y_TEST, predicted)
print(f"Accuracy: {accuracy}")
```

```python
def filter_news_by_category(docs, categories, selected_category):
    """
    Filter and display news articles based on the selected category.

    Parameters:
    docs (list): List of news articles as strings.
    categories (list): List of predicted categories corresponding to the docs.
    selected_category (str): The category to filter news articles by.

    Returns:
    list: List of news articles belonging to the selected category.
    """
    filtered_news = [doc for doc, category in zip(docs, categories) if category == selected_category]
    return filtered_news
```
